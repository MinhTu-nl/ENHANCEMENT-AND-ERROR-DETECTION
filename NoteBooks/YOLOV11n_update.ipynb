```python
# --- 1. C√ÄI ƒê·∫∂T TH∆Ø VI·ªÜN ---
# Ch·∫°y n·∫øu b·∫°n ch∆∞a c√†i ultralytics
!pip install ultralytics

# --- 2. IMPORT C√ÅC TH∆Ø VI·ªÜN C·∫¶N THI·∫æT ---
import os
import yaml
from ultralytics import YOLO
from pathlib import Path
```

```python
# --- 3. THI·∫æT L·∫¨P TH√îNG S·ªê HU·∫§N LUY·ªÜN ---
# B·∫°n c√≥ th·ªÉ ch·ªânh l·∫°i c√°c gi√° tr·ªã n√†y theo nhu c·∫ßu
data_path = 'drive/MyDrive/defects/dataset.yaml'
weights_path = 'yolov5s.pt'
imgsz = 640
batch_size = 32
epochs = 150
project = 'drive/MyDrive/weights/YoloV5'
name = 'defect_detection_exp'
device = ''  # '' = auto, ho·∫∑c 'cpu', '0', '0,1', ...
```

```python
# --- 4. H√ÄM KI·ªÇM TRA V√Ä C·∫¨P NH·∫¨T ƒê∆Ø·ªúNG D·∫™N TRONG FILE YAML ---
def check_data_path(data_yaml_path):
    try:
        with open(data_yaml_path, 'r') as f:
            data = yaml.safe_load(f)

        if 'path' in data and not os.path.isabs(data['path']):
            abs_path = os.path.abspath(os.path.join(os.path.dirname(data_yaml_path), data['path']))
            print(f"Updating relative path in {data_yaml_path} to absolute path:\n {abs_path}")
            data['path'] = abs_path
            with open(data_yaml_path, 'w') as f:
                yaml.dump(data, f, sort_keys=False)
        elif 'path' not in data:
            print(f"Warning: 'path' key not found in {data_yaml_path}.")
    except Exception as e:
        print(f"Error processing {data_yaml_path}: {e}")

# G·ªçi h√†m c·∫≠p nh·∫≠t ƒë∆∞·ªùng d·∫´n
check_data_path(data_path)

# --- 5. T·∫†O M√î H√åNH T·ª™ TR·ªåNG S·ªê --
# T·ª± ƒë·ªông t·∫£i yolov5s n·∫øu ch∆∞a c√≥
model = YOLO("yolov5s.pt")

# --- 6. IN RA C·∫§U H√åNH ƒêANG S·ª¨ D·ª§NG ---
print(f"Starting training with the following parameters:")
print(f" > Data YAML: {data_path}")
print(f" > Weights: {weights_path}")
print(f" > Image Size: {imgsz}")
print(f" > Batch Size: {batch_size}")
print(f" > Epochs: {epochs}")
print(f" > Project: {project}")
print(f" > Name: {name}")
print(f" > Device: {'cuda' if device == '' or 'cuda' in device else 'cpu'}")
```

```python
# --- 7. HU·∫§N LUY·ªÜN M√î H√åNH ---
try:
    results = model.train(
        data=data_path,
        imgsz=imgsz,
        epochs=epochs,
        batch=batch_size,
        project=project,
        name=name,
        device=device if device else None,
        exist_ok=False
    )
    print("Training completed successfully.")
    print(f"Results saved to: {results.save_dir}")
except Exception as e:
    print(f"An error occurred during training: {e}")
```
